```python
class No():
    def __init__(self, nome, valor):
        self.nome = nome
        self.valor = valor
        self.proximo = None
        
class ListaDesejos:
    def __init__(self):
        self.inicio = None #LISTA COMEÇA VAZIA

    def adicionar_item(self, nome, valor):
        novo_no = No(nome, valor)
        if self.inicio is None:
            self.inicio = novo_no
        else:
            atual = self.inicio
            while atual.proximo is not None:
                atual = atual.proximo
            atual.proximo = novo_no
    def calcular_total(self):
        total = 0.0
        atual = self.inicio
        while atual is not None:
            total += atual.valor
            atual = atual.proximo
        return total

# Cria a lista de desejos
minha_lista = ListaDesejos()

# Adiciona os itens conforme o exemplo do PDF [cite: 15]
minha_lista.adicionar_item("Fone de ouvido", 150.00)
minha_lista.adicionar_item("Mouse", 80.00)
minha_lista.adicionar_item("Teclado", 120.00)

# Calcula o total [cite: 8]
total_compra = minha_lista.calcular_total()

# Imprime o resultado formatado 
print(f"Valor total da compra: R$ {total_compra:.2f}")
        
```
